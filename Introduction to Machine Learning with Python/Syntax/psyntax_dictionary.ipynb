**Question 3**

```python
# Use: This program converts two Python lists into a dictionary where elements from the first list become keys and elements from the second list become values.
# Input: 
# Output:

keys = ['Ten', 'Twenty', 'Thirty']
values = [10, 20, 30]

def q3_fn(kys, vals):
    my_dict = {}
    for i in range(len(kys)):
        my_dict[kys[i]] = vals[i]
        
    return my_dict

def q3_fn_v2(kys, vals):
    return dict(zip(kys, vals))

def q3_fn_v3(kys, vals):
    my_dict = {}
    n = len(kys)
    
    for i in range(n):
        my_dict.update({kys[i]: vals[i]})

    return my_dict

print(q3_fn_v3(keys, values))
```

**Question 6**

```python
# Use: This program counts character frequencies.
# Input: 
# Output:

str1 = "Johannesburg"

def q6_fn(s):
    my_dict = {}
    for char in s:
        if char in my_dict:
            my_dict[char] += 1
        else:
            my_dict[char] = 1
    
    return my_dict

print(q6_fn(str1))
```

**Question 12**

```python
# Use: Delete a list of keys from a dictionary 
# Input: 
# Output:

sample_dict = {
    "name": "Kelly",
    "age": 25,
    "salary": 8000,
    "city": "New York"
}

keys = ["name", "salary"]

def q12_fn(dicti, kys):
    for key in kys:
        if key in dicti:
            dicti.pop(key)
    
    return dicti

print(q12_fn(sample_dict, keys))
```

**Question 15**

```python
# Use: This program obtain the minimum value of a key within a dictionary
# Input:
# Output:

sample_dict = {
  'Physics': 82,
  'Math': 65,
  'history': 75
}

def q15_fn(dicti):
    result = ''
    min = 10000000
    for i in sample_dict.values():
        if(i < min):
            min = i
    for k in dicti:
        if dicti[k] == min:
            result = k 

    return result  

print(q15_fn(sample_dict))


def q15_fn_v2(dicti):
    return min(dicti, key=dicti.get)

print(q15_fn_v2(sample_dict))
```

```python
# Use: This program inverts the items within a dictionary.
# Input:
# Output:

original_dict = {'a': 1, 'b': 2, 'c': 3}

def q17_fn(input_dict):
    inv_dict = {}
    for k, v in input_dict.items():
        inv_dict[v] = k
    
    return inv_dict

print(q17_fn(original_dict))
```
