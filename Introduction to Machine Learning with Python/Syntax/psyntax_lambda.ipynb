```python
# Use: This program uses a lambda function to add 15 to a given number which is passed as an argument and also creates a lambda function that multiplies argument x with argument y and prints the result.
# Input:
# Output:

a = 20
b = 20
def q1_fn(n, z):
    add_int = lambda x: x + 15
    mmply_arg  = lambda x, y: x * y  

    return add_int(n), mmply_arg(n, z)

res1, res2 = q1_fn(a, b)

print(res1, res2)
```

**Question 16**

```python
# Use: Present second lowest value within a dictionary
# Input:
# Output:

students = []
second_name = []
second_low = 0
n = int(input("Input number of students: "))

for _ in range(n):
    s_name = input("Name: ")
    score = float(input("Grade: "))
    students.append([s_name, score])


print("\nNames and Grades of all students: ")
print(students)

def q16_fn(lst):
    # Sort in order
    order = sorted(lst, key = lambda x: int(x[1]))
    # Find the second lowest grade among students
    for i in range(n):
        if order[i][1] != order[0][1]:
            second_low = order[i][1]
            break
    
    return second_low

print("Second lowest: {}".format(q16_fn(students)))
```

```python
# Use: Palindrome Filter Lambda
# Input:
# Output:

list1 = ['php', 'w3r', 'Python', 'abcd', 'Java', 'aaa']

def q18_fn(lst):
    palindrome = list(filter(lambda x: x if x[::] == x[::-1] else '', lst))

    return palindrome

print(q18_fn(list1))

def q18_fn_v2(lst):
    result = list(filter(lambda x: (x == "".join(reversed(x))), list1))

    return result

res = q18_fn_v2(list1)
print(res)
```

```python
# Use: Positive Negative Sum Lambda
# Input:
# Output:

list1 =  [2, 4, -6, -9, 11, -12, 14, -5, 17]

def q23_fn(lst):
    total_positive = sum(list(filter(lambda x: x > 0, lst)))
    total_negative = sum(list(filter(lambda x: x < 0, lst)))

    return total_positive, total_negative


sum_positive, sum_negative = q23_fn(list1)

print(f"Sum of the positive numbers: {sum_positive}\nSum of negative numbers: {sum_negative}")
```
