```python
# Use: Creates a string made of input string's first, second and last characters
# Input: s: Any string
# Output: A string made out of 1, 2, nth character

s1 = "Shaqueefa" 

def q1a_fn(s):
    new_str = ""
    for i in range(0, len(s)):
        if(i == 0 or i == 1 or i == len(s)-1):
            new_str+=s[i]

    return new_str

print(q1a_fn(s1))


```

```python
# Use: Arrange a string characters such that lowercase letters should come first
# Input:
# Output:

s1 = "PyNaTive"

def q4_fn(s):
    lower = ""
    upper = ""
    for char in s:
        if(char.isupper()):
            upper+=char
        else:
            lower+=char
    
    return lower+upper

print(q4_fn(s1))
```

```python
# Use: Count all letters, digits, and special symbols from a given string
# Input: 
# Output:

s1 = "P@#yn26at^&i5ve"

def q5_fn(s):
    l = 0
    d = 0
    z = 0
    for char in s:
        if(char.isalpha()):
            l+=1
        elif(char.isdigit()):
            d+=1
        else:
            z+=1
    
    return (l, d, z)

print(q5_fn(s1))


```

```python
# Use: Given two strings, s1 and s2, this program creates s3 made of the first character of s1, the last character of s2. Next, second char of s1 and second to last char of s2.
# Input:
# Output:

s1 = "Abc"
s2 = "Xyz"
vision = ['A', 'z',
          'b', 'y',
          'c', 'X']

# s1 = print in order
# s2 = print in reverse order
# s3 = add s1+s2 with that logic in mind

def q6_fn(s1, s2):
    result = ""
    n = min(len(s1), len(s2))

    for i in range(n):
        result += s1[i] + s2[len(s2) - 1 - i]

    if len(s1) > n:
        result += s1[n:]
    if len(s2) > n:
        result += s2[:len(s2) - n]

    return result

print(q6_fn(s1, s2))
```

**Question 9**

```python
# Use: Calculate the sum and average of the digits present in a string.
# Input:
# Output:

s1 = "PYnative29@#8496"

def q9_fn(s):
    d = []
    for char in s:
        if(char.isdigit()):
            d.append(int(char))
    
    total = sum(d)
    avg = round(total/len(d), 2)
    
    return total, avg

sums, average = q9_fn(s1) 

print("The Sum: {}\nThe Average: {}".format(sums, average))
```

**Question 14**

```python
# Use: Remove empty strings from a list of strings.
# Input: lst: A list containing different strings.
# Output: Updated list without empty strings

str_lst = ["Emma", "Jon", "", "Kelly", None, "Eric", ""]

def q14_fn(lst):
    result = []
    for word in lst:
        if not word:
            continue
        else:
            result.append(word)
            
    return result

q14_fn(str_lst)

def q14_fn_v2(lst):
    return list(filter(None, lst))

q14_fn_v2(str_lst)
```
