**Question 5**

```python
# Use: Displays numbers from a list using a loop
# Input: A list of any numbers
# Output: Numbers from the list that satisfy all conditions
# Conditions:
#             1. The number must be divisible by five; 
#             2. If the number is greater than 150, then skip it and move to the following number;
#             3. If the number is greater than 500, then stop the loop;

nums = [25, 11, 13, 14, 17, 20, 46, 24, 53, 63, 345, 142, 125, 341, 22, 13, 53, 64, 501, 101]

def q5_fn(num_lst):
    res_lst = []
    
    for n in range(len(num_lst)):
        if(num_lst[n] > 150):
            continue
        if(num_lst[n] > 500):
            break
        if(num_lst[n] % 5 == 0):
            res_lst.append(num_lst[n])
        
    return res_lst

print(q5_fn(nums))


```

**Question 11**

```python
# Use: Returns all prime numbers within a range 
# Input: p_range: A range of prime numbers
# Output: Prime numbers within that range

s = 25
e = 100

def is_prime(x):
    if(x == 1 or x == 2):
        return False
    for i in range(2, x, 1):
        if(x % i == 0):
            return False
    return True

def q11_fn(start, end):
    primes = []
    for i in range(start, end):
        if(is_prime(i)):
            primes.append(i)
    
    return primes

print(q11_fn(s, e))
```
