**Question 5**

```python
# Use: Symmetric Difference. Returns a new set containing elements that are unique to either set1 or set 2, but not in both. Like finding the union and then removing the intersection
# Input:
# Output:

set1 = {10, 20, 30, 40, 50}
set2 = {30, 40, 50, 60, 70}

sym_diff_set = set1.symmetric_difference(set2)

print(sym_diff_set)
```

**Question 16**

```python
# Use: Write code to count the number of unique words given a sentence.
# Input: s: Any sentence
# Output: Number of unique words in that sentence

sentence = "A dog is a simple animal and a selfless animal"  

def q16_fn(s):
    return len(set(s.lower().split()))

q16_fn(sentence)
```
