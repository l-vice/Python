```python
xarr = [1, 3, 4, 6, 7, 8, 10, 11, 12, 14, 15]
yarr = [2, 4, 6, 7, 8, 11, 12, 13, 15]

# -------------------------------------------------------------------------------
# Use: Find the intersection between arrays using a lambda function

def array_intersection(arr1, arr2):
    return list(filter(lambda x: x in arr1, arr2))
 
zarr = array_intersection(xarr, yarr)

# -------------------------------------------------------------------------------
# Use: Find the union between two arrays


def array_union(arr1, arr2):
    return list((lambda x, y: set(x) | set(y))(arr1, arr2))

warr = array_union(xarr, yarr)

# -------------------------------------------------------------------------------
# Use: Find the difference of arrays

def array_difference(arr1, arr2):
    return list(filter(lambda x: x not in arr2, arr1))

varr = array_difference(xarr, yarr)

# -------------------------------------------------------------------------------
# Use: Find the asymmetric difference of arrays
# Symmetric Difference: A set with elements in either set or other but not both.

def array_symmetric_difference(arr1, arr2):
    return list((lambda x, y: set(x).symmetric_difference(set(y)))(arr1, arr2))

uarr = array_symmetric_difference(xarr, yarr)

# -------------------------------------------------------------------------------
# Use: Find the common elements between two arrays

def array_common_elements(arr1, arr2):
    return list((lambda x, y: set(x).intersection(set(y)))(arr1, arr2))

tarr = array_common_elements(xarr, yarr)

# -------------------------------------------------------------------------------
# Output Result

print(f"Initial Arrays ---> Array 1: {xarr} | Array 2: {yarr}\nOuput -->\nArray intersection: {zarr}\nArray Union: {warr}\nArray difference: {varr}\nArray Symmetric Difference: {uarr}\nArray Common Elements: {tarr}")
```

```python
my_list = ['bcda', 'abce', 'cbda', 'cbea', 'adcb']

from collections import defaultdict
# -------------------------------------------------------------------------------
# Use: Write a Python program to identify the word with the most anagrams in a given list using lambda.

def max_anagram(lst):
    grouped_anagrams = defaultdict(list)
    for word in lst:
        sorted_word = ''.join(sorted(word))
        grouped_anagrams[sorted_word].append(word)
    
    max_group = max(grouped_anagrams.values(), key = lambda x: len(x))

    return max_group[0]

result = max_anagram(my_list)
print(result)
```

```python
my_list = ['bcda', 'abce', 'cbda', 'cbea', 'adcb']

from collections import Counter
# -------------------------------------------------------------------------------
# Use: Write a Python program to filter out words that do not have any anagrams in the list using lambda.

def not_anagram(lst):
    return list(filter(lambda x: any(Counter(x) == Counter(other) and x != other for other in lst), lst))

result = not_anagram(my_list)
print(result)
```

```python
# Q3
# Use: Write a Pthon program to identify the word with the most anagrams in a give list using lambda.
from collections import defaultdict

my_list = ['qwerty', 'listen', 'silent', 'cat', 'bar', 'stereotype', 'wigahoo', 'persia']

def find_anagram(lst):
    grouped_anagrams = defaultdict(list)
    for word in lst:
        sorted_word = ''.join(sorted(word))
        grouped_anagrams[sorted_word].append(word)
    
    max_group = max(grouped_anagrams.values(), key = lambda x: len(x))

    return max_group[0]


result = find_anagram(my_list)
result
```

```python
# 93
# Use: Write a Python program to identify the word with the most anagrams in a given list using lambda.
from collections import defaultdict
my_list = ["glue", "true", "silent", "car", "arc", "PhP", 'qqqqqq']

def find_max(lst):
    grouped_anagrams = defaultdict(list)
    for word in lst:
        sorted_word = ''.join(sorted(word))
        grouped_anagrams[sorted_word].append(word)
    
    max_group = max(grouped_anagrams.values(), key = lambda x: len(x))

    return max_group[0]

result = find_max(my_list)
result 
```
