```python
# Use: Check Palindrome Number. A palindrome number reads the same forwards and backwards. For example, 545 is a palindrome number.
# Input: 
# Output:

def q9_fn(num):
    nmbr = str(num)
    n = len(nmbr)
    # FORWARD
    forward = ""
    for f in range(n):
        forward +=nmbr[f]
    # BACKWARD
    backward = ""
    for b in range(n - 1, -1, -1):
        backward +=nmbr[b]
    
    if(forward == backward):
        return print("{} is a palindrome number".format(num))
    else:
        return print("{} is not a palindrome number".format(num))


num = 455

q9_fn(num)
```

```python
# Use: Given two lists of numbers, write code to create a new list containing odd numbers from the first list and even numbers from the second list.
# Input: List A, B
# Output: List C, containing odd numbers from A, and even from B.

A = [10, 20, 25, 30, 35]
B = [40, 45, 60, 75, 90]

def q10_fn(list1, list2):
    result = []
    n = len(list1)
    for i in range(n):
        #  LST1 // ODD VALUES
        if(list1[i] % 2 != 0):
            result.append(list1[i])
        # LST2 // EVEN VALUES
        if(list2[i] % 2 == 0):
            result.append(list2[i])
    
    return result

C = q10_fn(A, B)

print(C)    
```

```python
# Use: Check if a user-entered string contains any digits using a for loop
# Input: string_name = A string
# Output: 

def q20_fn():
    string_name = str(input("Enter a string: "))
    count = 0
    for i in range(len(string_name)):
        if(string_name[i].isdigit() == True):
            count+=1
            if(count > 0):
                print("The string contains at least one digit.")
        else:
            count+=1
            if(count == len(string_name)):
                print("The string does not contain a digit")
    
    return ""

print(q20_fn())
```

```python
def contains_digits(text):
    for char in text:
        if '0' <= char <= '9':
            return True
    return False

user_input = input("Enter a string: ")

if contains_digits(user_input):
    print("The string contains at least one digit.")
else:
    print("The string does not contain any digits")
```

```python
# Use: Capitalize the word of each letter in a string.
# Input: string_name = A string 
# Output: Updated string_name with capitalized first letter of each word

string_name = "shaqueefa is a very nice girl"

def q21_fn(string_name):
    cap_words = []

    words = string_name.split()
    for word in words:
        word = word.capitalize()
        cap_words.append(word)
        
        sentence = ' '.join(cap_words)

    return sentence

q21_fn(string_name)


```

```python
# Use: This function creates a simple countdown timer using a while loop.
# Input: time: Time limit.
# Output: Countdown for n number of seconds.

from time import sleep

duration = 20

def q22_fn(dur):
    i = 0
    while True: 
        
        print("Time remaining: {} seconds".format(dur-i))
        sleep(1)
        i+=1
        if(i == dur):
            break
    
    return "Time's up!"

print(q22_fn(duration))
```
