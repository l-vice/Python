```python
# Use: Access Nested Tuples
# Input: 
# Output:

tpl = ("Orange", [10, 20, 30], (5, 15, 25)) 

def q5_fn(tup):
    target = 0
    for i in range(len(tup)):
        if isinstance(tup[i], list):
            for j in tup[i]:
                if(j == 20):
                    target = j
    
    return target

print("Long way: {}\n".format(q5_fn(tpl)))

# OR 

print("Short way: {}\n".format(tpl[1][1]))
```

**Question 11**

```python
# Use: Function Returning Tuple with min and max values within a list of numbers
# Input: lst: A list
# Output: tuple object with min and max values

import sys

nums = []

def q11_fn(lst):
    if not lst:
        return tuple((None, None))
    min = sys.maxsize
    max = 0
    for i in range(len(lst)):
        if(lst[i] > max):
            max = lst[i]
        if(lst[i] < min):
            min = lst[i]
    
    return tuple((min, max))

print(q11_fn(nums))
    
```

```python
# Use: Comparing Tuples
# Input: t1: Tuple 1; t2: Tuple 2
# Output: Compares the size of tuples

t1 = (1, 2, 3)
t2 = (1, 2, 4)

def q12_fn(t1, t2):
    if(t1 > t2):
        return print("{} is bigger than {}".format(t1, t2))
    if(t1 < t2):
        return print("{} is bigger than {}".format(t2, t1))
    else:
        return print("{} is equal to {}".format(t1, t2))

print(q12_fn(t1, t2))
```

**Question 17**

```python
# Use: Sorting a tuple of nested tuples by 2nd item
# Input:
# Output:

t = (('a', 23), ('b', 37), ('c', 11), ('d', 29))

def q17_fn(tpl):
    tpl = list(tpl)

    tpl.sort(key=lambda x: x[1])

    return tuple(tpl)

print(q17_fn(t))

```

**Question 18**

```python
# Use: This function counts the number of occurences of item 50 from a tuple.
# Input: t1: A tuple; x: A target number
# Output:

t1 = (50, 10, 60, 70, 50)
n = 50

def q18_fn(tpl, x):
    target = x
    count = 0
    for i in range(len(tpl)):
        if(tpl[i] == target):
            count+=1
    
    return count

print(q18_fn(t1, n))

def q18_fn_v2(tpl, x):
    return tpl.count(x)

print(q18_fn_v2(t1, n))
```

```python
# Use: This function checks whether all items inside a tuple are the same
# Input:
# Output:

t1 = (45, 45, 45, 45)

def q19_fn(tpl):
    n = len(tpl)
    total = sum(tpl)
    if(t1[0] == total/n):
        return True
    
    return False

def q19_fn_v2(tpl):
    return all(i == tpl[0] for i in tpl)

print(q19_fn_v2(t1))
```
