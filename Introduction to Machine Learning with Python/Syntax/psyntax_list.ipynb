**Question 3**

```python
# Use: Returns the sum and average of all elements in a list
# Input:
# Output:

my_list = [10, 20, 30, 40, 50]

def q3_fn(lst):
    lst_sum = 0
    lst_avg = 0

    for i in lst:
        lst_sum+=i
        
    lst_avg = lst_sum/len(lst)

    return lst_sum, lst_avg

total, average = q3_fn(my_list)

print("The Sum: {}\n The Average: {}".format(total, average))
```

**Question 4**

```python
# Use: Reverse a list
# Input:
# Output:

list1 = [100, 200, 300, 400, 500]

def q4_fn(lst):
    rev_lst = []
    n = len(lst)
    for i in range(n-1, -1, -1):
        rev_lst.append(lst[i])
    
    return rev_lst

def q4_fn_v2(lst):
    lst.reverse()
    return lst

q4_fn_v2(list1)
```

**Question 13**

```python
# Use: Remove all occurences of a specific item in a list
# Input:
# Output:

t = 5

list1 = [5, 20, 15, 20, 25, 50, 20]

def q13_fn(lst, t):
    for i in lst:
        if i == t:
            lst.remove(i)

    return lst

def q13_fn_v2(lst, t):
    return [i for i in lst if i != t]

q13_fn_v2(list1, t)
```

**Question 17**

```python
# Use: Concatenate two lists index-wise.
# Input:
# Output:

list1 = ["M", "na", "i", "Ke"]
list2 = ["y", "me", "s", "lly"]

def q17_fn(lst1, lst2):
    new_list = []
    n = len(lst1)
    for i in range(n):
        new_list.append(lst1[i]+lst2[i])
    
    return new_list

q17_fn(list1, list2)

def q17_fn_v2(lst1 , lst2):
    return [i+j for i, j in zip(lst1, lst2)]
    
q17_fn_v2(list1,list2)

```

**Question 18**

```python
# Use: Concatenate two lists in the following order.
# Input:
# Output:

list1 = ["Hello", "take"]
list2 = ["Dear", "Sir"]

def q18_fn(lst1, lst2):
    new_lst = []
    for i in lst2:
        new_lst.append(lst1[0]+' '+i)
        new_lst.append(lst1[1]+' '+i)
        
    return new_lst

q18_fn(list1, list2)

def q18_fn_v2(lst1, lst2):
    return [x + ' ' + y for x in lst1 for y in lst2]
```

**Question 22**

```python
# Use: Replace list's item with new value if found.
# Input:
# Output:

elem = 50
list1 = [5, 10, 50, 20, 25, 50, 20]


def q22_fn(lst, ele, repl):
    for i in range(len(lst)):
        if lst[i] == ele:
            lst[i] = repl
            return lst

def q22_fn_v2(lst, ele, repl):
    index = lst.index(ele)
    
    lst[index] = repl

    return lst

q22_fn_v2(list1, 50, 200)
```
